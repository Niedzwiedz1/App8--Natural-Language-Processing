```python
with open("miracle_in_the_andes.txt","r") as file:
    book = file.read()
```

```python
import re
```

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book)
findings
```

```python
d = {}
for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d_list = [(value, item) for (item, value) in d.items()]
d_list = sorted(d_list, reverse=True)
```

```python
from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")
```

```python
filtered_words = []
for number, word in d_list:
    if word not in english_stopwords:
        filtered_words.append((number, word))
        
```

# Sentiment Analysis: What is the most positive and the negative chapter

### Chapter Sentiment analysis

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
analyzer.polarity_scores("Love is Love, I want to love you")
```

```python
import re
```

```python
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
```

```python
chapters = chapters[1:]
```

```python
for index, chapter in enumerate(chapters):
    score = analyzer.polarity_scores(chapter)
    print(score["neg"], score["pos"])
    if score["neg"]>score["pos"]:
        print(f"Chapter {index+1} is negative")
    else:
        print(f"Chapter {index+1} is positive")
```
